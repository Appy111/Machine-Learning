```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
cat_col=x.select_dtypes(object).columns
x[cat_col]=oe.fit_transform(x[cat_col])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
logreg=LogisticRegression()
knn=KNeighborsClassifier()
svm=SVC()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print(classification_report(ytest,ypred))
    return model
```

```python
mymodel(logreg)
```

```python
mymodel(knn)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ytest,ypred))
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
train=lr.score(xtrain,ytrain)
test=lr.score(xtest,ytest)
print(f'Training accuracy:{train} \n Testing Accuracy:{test}')
```

Best fit model as training & testing errors both are low -->
if model overfits --> apply regularization/cross validation

```python
mymodel(svm)
```

```python
mymodel(dt)
```

```python
dtr=DecisionTreeClassifier()
dtr.fit(xtrain,ytrain)
ypred=dtr.predict(xtest)

train=dtr.score(xtrain,ytrain)
test=dtr.score(xtest,ytest)
print(f'Training accuracy:{train} \n Testing Accuracy:{test}')
```

###### Solve overfitting

```python
for i in range (1,31):
    dt1=DecisionTreeClassifier(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)

    from sklearn.metrics import accuracy_score
    ac=accuracy_score(ytest,ypred)
    print(f'max_depth: {i} \n accuracy:{ac}')
```

```python
dt2=DecisionTreeClassifier(max_depth=13)
mymodel(dt2)

train=dt2.score(xtrain,ytrain)
test=dt2.score(xtest,ytest)
print(f'Training accuracy:{train} \n Testing Accuracy:{test}')
```
